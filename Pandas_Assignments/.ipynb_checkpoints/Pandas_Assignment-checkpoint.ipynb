```python
#2. Write a Pandas program to convert a Panda module Series to Python list and itâ€™s type.
s1 = pd.Series([10, 20, 30, 40, 50, 60, 70])
print(f"Series: {s1}")
print(f"Type = {type(s1)}")
l1 = s1.tolist()
print(f"List: {l1}")
print(f"Type = {type(l1)}")
```

```python
#3. Write a Pandas program to add, subtract, multiple and divide two Pandas Series.
   #Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]
s1 = pd.Series([2,4,6,8,10])
s2 = pd.Series([1, 3, 5, 7, 9])
op = s1 + s2
print(f"Addition = {op}")
op = s1 - s2
print(f"Subtract = {op}")
op = s1 * s2
print(f"Multiply = {op}")
op = s1 / s2
print(f"Divide = {op}")
```

```python
#4. Write a Pandas program to convert a dictionary to a Pandas series.
student = {
    "rollno" : 12,
    "name" : "Manish Zine",
    "Address" : "Buldhana",
    "Marks" : [98, 96, 92, 97, 94]
}
print(f"Dictionary: {student}")
print(f"Type = {type(student)}")
s1 = pd.Series(student)
print(f"Series: {s1}")
print(f"Type = {type(s1)}")
```

```python
#5. Write a Pandas program to convert a NumPy array to a Pandas series.
arr = np.array([1,2,3,4,5])
print(arr)
print(f"{type(arr)}")
s1 = pd.Series(arr)
print(s1)
print(f"{type(s1)}")
```

```python
#6. Write a Pandas program to change the data type of given a column or a Series.
s1 = pd.Series([10, 20, 30, 40, 50])
print(f"Series: {s1}")
print(f"Type: {s1.dtypes}")
s1 = s1.astype(float)
print(f"After change: {s1}")
print(f"Type: {s1.dtypes}")
```

```python
#7. Write a Pandas program to convert a given Series to an array.
s1 = pd.Series([10, 20, 30, 40, 50])
print(f"Series: {s1}")
s1 = s1.values
print(f"Series to array: {s1}")
print(f"Type: {type(s1)}")
```

```python
#8. Write a Pandas program to create a dataframe from a dictionary and display it.
student = {
    "rollno" : 12,
    "name" : "Manish Zine",
    "Address" : "Buldhana",
    "Marks" : [98, 96, 92, 97, 94]
}
print(f"Dictionary: {student}")
student = pd.DataFrame(student)
print(f"Data frame: {student}")
```
